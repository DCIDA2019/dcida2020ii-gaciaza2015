```python
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import corner
import emcee
```

```python
data = np.load('test_line.npy')                                 #Cargamos el archivo de datos
n = len(data[0])                                                           
data[2] = abs(data[2])
```

```python
plt.rcParams['figure.figsize'] = (8,4)
plt.errorbar(data[0], data[1], yerr=data[2], xerr = 0, fmt='o', color="black")
#plt.plot(data[0], -0.989*data[0]+4.633)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
def prior(u):
    if 10 < u[0] > 35 and 150 < u[1] > 230:
        prio = 0    #-np.log((35-(10))*(230-(150)))
    else:
        prio = -99999
    return prio
```

```python
def posterior(w):
    llt = 0
    ll = -(n/2)*np.log(2*np.pi)
    for i in range (n):
        fx = data[0,i]*w[0] + w[1]                                     #Aquí va el modelo
        llt+= -(n/2)*np.log(data[2,i])-(1/(2*data[2,i]))*(data[1,i]-fx)**2
    ll += llt + prior(w)
    return ll
```

```python
pos = [25,204] + 1e-4 * np.random.randn(12, 2)
nwalkers, ndim = pos.shape
sampler = emcee.EnsembleSampler(nwalkers, ndim, posterior)
sampler.run_mcmc(pos, 4000);
```

```python
fig, axes = plt.subplots(2, figsize=(8, 5), sharex=True)
samples = sampler.get_chain()
labels = ["m", "b"]

ax = axes[0]
ax.plot(samples[:, :, 0], "k", alpha=0.3)
ax.set_xlim(0, len(samples))
ax.set_ylim(17, 31)
ax.set_ylabel(labels[0])
ax.yaxis.set_label_coords(-0.1, 0.5)

ax = axes[1]
ax.plot(samples[:, :, 1], "k", alpha=0.3)
ax.set_xlim(0, len(samples))
ax.set_ylabel(labels[1])
ax.set_ylim(196, 212)
ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

```python
flat_samples = sampler.get_chain(discard=100, thin=15, flat=True)    #flat=True
print(flat_samples.shape)
```

```python
fig = corner.corner(
    flat_samples,
    range = [[20,27],[200,208]],
    labels=[r"$a$", r"$b$"],
    quantiles=[0.16, 0.5, 0.84],
    title_fmt='.4f',
    show_titles=True, title_kwargs={"fontsize": 12}
);
```

```python
from IPython.display import display, Math
for i in range(ndim):
    mcmc = np.percentile(flat_samples[:, i], [16, 50, 84])
    q = np.diff(mcmc)
    txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
    txt = txt.format(mcmc[1], q[0], q[1], labels[i])
    display(Math(txt))
```

```python
flat_samples = sampler.get_chain(discard=100, thin=15)    #flat=True
print(flat_samples.shape)
```

```python
s,me,R = np.zeros((12,ndim)),np.zeros((12,ndim)),np.zeros((ndim))
for i in range(12):
    s[i],me[i] = np.var(flat_samples[i],axis=0,ddof=1),np.mean(flat_samples[i],axis=0)
st = np.mean(s**2,axis=0)
for j in range(ndim):
    R[j] = np.sqrt(((((5000-1)/5000)*st[j])+ np.var(me,axis=0,ddof=1)[j])/st[j])

print(' R = ',R,'\n\n','Y las medias para cada caminante fueron:','\n',me)
```
