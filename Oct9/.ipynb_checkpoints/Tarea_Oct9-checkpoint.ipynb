## 1 - Find the cumulative distribution function for the Gaussian Distribution, and reproduce the plots. Choose a random number between 0 and 1, and use the CDF to assign the corresponding value of H0. Generate as many as you want, and make the histogram of H0 to verify you did it right. Use a mean of 70 and a sigma=2.

#### A continuación graficaremos el CDF gaussiano:
# $\frac{1}{\sigma\sqrt{2 \pi}}\int_{-\infty}^{0}{e^{-\frac{(x-\mu)^2}{2\sigma^2}}}dx = $
# $\phi(x|\mu,\sigma) = \frac{1}{2}\left[ er\!f\!\left( \frac{(x-\mu)^2}{\sqrt{2\sigma^2}}\right)+1\right]$

```python
def CDF(s, m, x):
    n = len(x)
    H = np.zeros(n)
    for j in range(n):
        H[j] = (1/2)*(sc.special.erf((x[j]-m)/(np.sqrt(2*s*s)))+1)
    return H
```

```python
#Función para gráficar
plt.style.use('dark_background')
#Esto es solo la función para graficar
#funcion(x,y,errores,vector p,pendiente,parámetro libre,error pendiente,error param. libre)
def scat(aj,h,plo,xa,xb,st):
    plt.rcParams['figure.figsize'] = (10,5)
    if plo == 'scatter':
        plt.scatter(aj,h,color='blue')
    elif plo == 'plot':
        plt.plot(aj,h,color='blue', linewidth=3)
    #plt.text(8,6,'$y = ax + b$',color='white')#,fontsize=15,bbox={'facecolor':'black','pad':10})
    plt.ylim(0, 1.1)
    plt.xlim(xa, xb)
    plt.xlabel('$H_0$')
    plt.ylabel(st)
    plt.grid()
    plt.show()
```

```python
h0 = np.arange(65,75,0.2)
n = len(h0)
#for i in range(n):
    #r[i] = np.random.rand()
fh = CDF(1,70,h0) #CumulativeDistributionFunction(sigma,media,H0)
```

```python
scat(h0,fh,'plot',66,74,'$f(H_0)$')
```

#### Ahora graficaremos un histograma dada cierta cantidad de números aleatorios y H0 asociados.

```python
hh0 = np.zeros(n)
pro = np.zeros(n)
for i in range(300000):
    r = np.random.rand()
    ind = np.argmin(abs(r-fh))
    pro[ind]+=1
pro=pro/max(pro)
```

```python
scat(h0,pro,'plot',66,74,'$P(H_0)$')
```

## 4 - Show explicitly that the linear least square method is derived from the minimization of the chisquare when the model is a straight line.
#### La función Chi-square para una linea recta $ax+b$ está definida como:
# $\chi^2 = \sum_i^n{(y_i - y(x_i,\theta))^2} = \sum_i^n{(y_i - ax_i - b)^2}$
#### Donde $\theta$ son los parámetros libres

#### Para minimizar la función hay que hacerlo respecto a los valores de a y b.
#### Entonces lo que buscamos es:
# $\frac{\partial \chi^2}{\partial a} = 0 \;, \; \frac{\partial \chi^2}{\partial b} = 0$
# $\frac{\partial \chi^2}{\partial a} = -2\sum_i^n{(y_i - ax_i - b)x_i} = 0$
# $a = \frac{\sum_i^n{y_ix_i}-b\sum_i^n{x_i}}{\sum_i^n{x_i^2}} \;\; (1)$
#### Mientras que para b:
# $\frac{\partial \chi^2}{\partial b} = -2\sum_i^n{(y_i - ax_i - b)} = 0$
# $nb = \sum_i^n{y_i}-a\sum_i^n{x_i} \;\; (2)$
#### Ahora sustituimos la ecuación 1 en la 2:
# $nb = \sum_i^n{y_i}-\sum_i^n{x_i}\left [ \frac{\sum_i^n{y_ix_i}-b\sum_i^n{x_i}}{\sum_i^n{x_i^2}}\right ] = \frac{\sum_i^n{y_i}\sum_i^n{x_i^2}-\sum_i^n{x_i}\sum_i^n{y_ix_i}+b(\sum_i^n{x_i})^2}{\sum_i^n{x_i^2}}$
# $\frac{b(n\sum_i^n{x_i^2}-(\sum_i^n{x_i})^2)}{\sum_i^n{x_i^2}} = \frac{\sum_i^n{y_i}\sum_i^n{x_i^2}-\sum_i^n{x_i}\sum_i^n{y_ix_i}}{\sum_i^n{x_i^2}}$
#### Entonces:
# $\boxed{b = \frac{\sum_i^n{x_i}\sum_i^n{x_iy_i}\ -\ \sum_i^n{y_i}\sum_i^n{x_i^2}}{(\sum_i^n{x_i})^2- \ n\sum_i^n{x_i^2}}}$
#### Ahora sustituimos este valor en 2, para obtener a:
# $n\frac{\sum_i^n{x_i}\sum_i^n{x_iy_i}\ -\ \sum_i^n{y_i}\sum_i^n{x_i^2}}{(\sum_i^n{x_i})^2- \ n\sum_i^n{x_i^2}} = \sum_i^n{y_i}-a\sum_i^n{x_i}$
# $a\sum_i^n{x_i} = \frac{-n\sum_i^n{x_i}\sum_i^n{x_iy_i}\ +\ n\sum_i^n{y_i}\sum_i^n{x_i^2}+ \sum_i^n{y_i}(\sum_i^n{x_i})^2 - n\sum_i^n{y_i}\sum_i^n{x_i^2}}{(\sum_i^n{x_i})^2- \ n\sum_i^n{x_i^2}}$
# $\boxed{a = \frac{\sum_i^n{y_i}\sum_i^n{x_i}-n\sum_i^n{x_iy_i}}{(\sum_i^n{x_i})^2- \ n\sum_i^n{x_i^2}}}$

## 5- Write down a Poisson likelihood, and identify which parameter(s) correspond to the data and which to the model.

## La distribución de poisson se define como:
## $f(k,\lambda) = \frac{e^{-\lambda}\lambda^k}{k!}$
### Donde $\lambda$ es un parámetro positivo que representa el número de veces que se espera que ocurra el fenómeno durante un intervalo dado y $k$ es el número de ocurrencias del evento o fenómeno (la función nos da la probabilidad de que el evento suceda precisamente k veces)
## Su Likelihood será entonces:
## $\mathbf{L} \; \propto \; \prod_i^n \frac{e^{-\lambda}\lambda^k_i}{k_i!}$
### Donde los $k_i$ provedrán de los datos y $\lambda$ será el modelo.
