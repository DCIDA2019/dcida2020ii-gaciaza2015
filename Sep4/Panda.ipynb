# Tablas con Panda y requests

```python
import requests
import pandas as pd
```

## Podemos crear tablas a partir de arreglos o diccionarios

```python
nom = {"Pablo": 7, "Laura": 4, "Sergio": 3, "Helen": 9}
tab = pd.Series(nom)
tab
```

## Podemos etiquetar y anexar datos nuevos 

```python
din = [234, 897, 6563, 3780]
tab = pd.DataFrame({'Edad':tab, 'Dinero':din})
tab
```

## Podemos ordenar los datos a conveniencia

```python
tab = tab.sort_values("Dinero", ascending = False)
tab
```

## Podemos extraer tablas de internet usando requests 

```python
url = "https://www.indexmundi.com/map/?t=0&v=21&r=xx&l=en"
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[1]
df[df.Population > 45000000]
```
