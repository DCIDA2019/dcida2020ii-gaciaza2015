# Tablas con Panda y requests

```python
import requests
import pandas as pd
```

## Podemos crear tablas a partir de arreglos o diccionarios

```python
nom = {"Pablo": 7, "Laura": 4, "Sergio": 3, "Helen": 9}
tab = pd.Series(nom)
tab
```

## Podemos etiquetar y anexar datos nuevos 

```python
din = [234, 897, 6563, 3780]
tab = pd.DataFrame({'Edad':tab, 'Dinero':din})
tab
```

## Podemos ser selectivos con los datos mostrados

```python
tab[tab.Dinero>1000]
```

## Podemos ordenar los datos a conveniencia

```python
tab = tab.sort_values("Dinero", ascending = False)
tab
```

## Podemos extraer tablas de internet usando requests 

```python
url = "https://www.indexmundi.com/map/?t=0&v=21&r=xx&l=en"
html = requests.get(url).content                                  #Este comando extrae la informaciÃ³n
df_list = pd.read_html(html)                                      #Panda se encarga de darle forma
df = df_list[1]                                                   #La tabla se encuentra en el index "1"
df[df.Population > 100000000]
```

## Podemos extraer datos de otra tabla y juntarlos en otra

## Igualmente podemos hacer operaciones con los datos

```python
url = "https://www.indexmundi.com/map/?t=0&v=2226&r=xx&l=en"                    #Leemos nuevo link
html = requests.get(url).content                                  
df_list = pd.read_html(html)                                      
df2 = df_list[1]                                                  
aux=[]

for i in range(198):                                         #Extraemos los datos que requerimos y los ordenamos
    x = False
    for j in range(188-i):
        if df["Country"][i]==df2["Country"][j]:
            aux.append(df2["Physicians density (physicians/1,000 population)"][j])
            x = True
    if x == False:
        aux.append(0)

#Con DataFrame seleccionamos los datos de las dos tablas que queremos mostrar y creamos una columna nueva
tab = pd.DataFrame({'Country':df["Country"], 'Poblacion': df["Population"], 'Fisicos(Por cada 1000)':aux, 'Fisicos': aux*df["Population"]/1000})
tab[tab.Fisicos> 100000]
```
