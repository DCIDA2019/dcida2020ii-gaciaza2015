## Método de Euler para péndulo simple. (Sin Numpy)
## $\theta^{\prime \prime} = -\frac{g}{l} \sin{\theta}$

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import math

h = 0.01
t = [0]                             #Tiempo
yo = [math.pi/4]                    #Angulo cambiante con pi/4 inicial
n = 1000                            #número de iteraciones
g = 9.80665
l = 1                               #Longitud del péndulo

wo = math.sqrt(g/l)
wi = wo

def euler(y,w):                     #Método de Euler
    wn = w - h*wo*wo*math.sin(y)    #Primera derivada
    yn = y + h*wn                   #Segunda derivada
    return yn, wn

for i in range(n):
    s, wi = euler(yo[i],wi)
    yo.append(s)
    t.append(t[i]+h)

    
plt.rcParams['figure.figsize'] = (12,4)
plt.plot(t, yo, color='black')
plt.xlabel('Tiempo')
plt.ylabel('$\Theta^{\prime \prime}$')
plt.grid()
plt.show()
```
